<u>**Ex2 :**</u> Il s'agit de gnrer un programme cod en python qui vous demande les coordonnes de deux points A(x1,y1) et B(x2,y2), qui calcule la distance AB

$$AB=\sqrt{{(x2 - x1)}^2 + {(y2 - y1)}^2}$$

   - Vous devez importer la fonction <i>"racine carre"</i> de la bibliothque(module) **math**, elle se nomme **sqrt()**
   - Les fonctions input et print seront utilises
   - la fonction "lever au carr" doit s'crire sous la forme suivante : 'x**2'
   - Les variables seront les suivantes : **X1, Y1, X2, Y2, AB**
   - Vous utiliserez les f-strings
   - Vous utiliserez le formatage des nombres en f-strings tel que nous ayons alignement  droite, le signe +, la largeur 10 chiffres, 2 chiffres significatifs de 2 chiffres aprs la virgule 



```python
from math import sqrt
X1=int(input("Entrez l'abscisse de A : "))
Y1=int(input("Entrez l'ordonne de A : "))
X2=int(input("Entrez l'abscisse de B : "))
Y2=int(input("Entrez l'ordonne de B : "))
AB=sqrt(((X2-X1)**2)+((Y2-Y1)**2))
print(f'La distance AB est {AB}')
print(f'La distance AB est {AB : >+.2f}')
       
```

<u>**Ex3 :**</u> Il s'agit de gnrer un trac de courbe en utilisant le module matplotlib et numpy

La courbe  tracer est la suivante :

$$y = e^{x}$$

En abscisse, x<sub>min</sub> = 0;    x<sub>max</sub> = 5


   - Vous devez importer le module <i>"pyplot"</i> de la bibliothque **mathplotlib** et nous renommerons le tout  **matplot**
   - Nous devrons galement importer la bibliothque de fonctions numpy
   - les fonctions **xlabel( ), ylabel( ), linspace( ), exp( )** seront ncessaires
   - le trac se fera en **pointills rouges**
   
   


```python
import matplotlib.pyplot as matplot
import numpy 
x=numpy.linspace(0,5,20)
matplot.xlabel("l'axe des abscisse (x) ")
matplot.ylabel("l'axe des ordonnes y=f(x) ")
matplot.plot(x,numpy.exp(x), 'r--')
matplot.grid(True)
```
